```python
import config
import openai
```

```python
openai.api_key = config.api_key
```

```python
def generate_text(prompt):
    response = openai.Completion.create(
        engine = "davinci-002",
        prompt = prompt,
        max_tokens = 10,
        temperature = 0.7
    )
    return response.choices[0].text.strip()
```

```python
prompt = "Once upon a time "
```

```python
result = generate_text(prompt)
```

```python
print(result)
```
