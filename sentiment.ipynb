```python
import pandas as pd
import numpy as np
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import transformers
from transformers import pipeline
```

```python
data = pd.read_csv("book_reviews_sample.csv")
```

```python
data.head()
```

```python
data['reviewText'][0]
```

```python
data['reviewText_clean'] = data['reviewText'].apply(lambda x: re.sub(r"[^\w\s]", "",x.lower()))
```

```python
data.head()
```

```python
vader_sentiment = SentimentIntensityAnalyzer()
```

```python
data['vader_sentiment_score'] = data['reviewText_clean'].apply(lambda review: vader_sentiment.polarity_scores(review)['compound'])
```

```python
data.head()
```

```python
bins = [-1, -0.1, 0.1, 1]
names = ['negative', 'neutral', 'positive']

data['vader_sentiment_label'] = pd.cut(data['vader_sentiment_score'], bins, labels=names)
```

```python
data.head()
```

```python
data['vader_sentiment_label'].value_counts().plot.bar()
```

```python
transformer_pipeline = pipeline('sentiment-analysis')
```

```python
transformer_labels = []

data['transformer_sentiment_label'] = data['reviewText_clean'].apply(lambda x: transformer_pipeline(x)[0]['label'])


```
